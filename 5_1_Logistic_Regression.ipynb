# 5-1 Wine Classification with Logistic Regression

```python
import pandas as pd

wine = pd.read_csv('wine.csv')
```

```python
wine.head()     # First 5 samles

# Forth columne: target -> 0: red wine, 1: white wine
```

```python
wine.info()

# Result: total 6497 samples
```

```python
wine.describe()

# describe(): mean, std, min, max
```

```python
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()  # numpy array
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
# Train set & test set
print(train_input.shape, test_input.shape)
```

```python
# StandardScaler class
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)        # Scaled data
test_scaled = ss.transform(test_input)          # Scaled data
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
# Coefficient & Intersection
print(lr.coef_, lr.intercept_)
```

## Decision Tree
- pp226

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()

# Root node/ leaf node
```

```python
# Control max_depth = 1 -> extend one node except the root node. 

plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()

# Gini impurity
```

## Branch

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)       # depth: 3
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
# Decision tree does not need to be pre-processed. 

dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(train_input, train_target)

print(dt.score(train_input, train_target))          # Not scaled
print(dt.score(test_input, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
print(dt.feature_importances_)      # alcohol, sugar, ph
```

## Example

```python
dt = DecisionTreeClassifier(min_impurity_decrease=0.0005, random_state=42)
dt.fit(train_input, train_target)

print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```
